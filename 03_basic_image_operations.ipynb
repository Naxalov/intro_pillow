```python
# Open image from file
flat_image = Image.open('flatuicolors.png')
# Show image using matplotlib
plt.imshow(flat_image)
plt.show()
# print image size
print(flat_image.size)
```

# Resize image using resampling
- image.NEAREST (Nearest neighbor interpolation)
- image.BILINEAR (Linear interpolation in a 2x2 environment)
- image.BICUBIC (Cubic spline interpolation in a 4x4 environment)
- image.LANCZOS (Lanczos interpolation in a 8x8 environment)
- image.HAMMING (Hamming interpolation in a 5x5 environment)
- image.BOX (Box interpolation (fastest))
 

```python
# Resize image using NEAREST resampling
img_NEAREST = flat_image.resize((500, 400),resample=Image.NEAREST)
img_BILINEAR = flat_image.resize((500, 400),resample=Image.BILINEAR)
img_BICUBIC = flat_image.resize((500, 400),resample=Image.BICUBIC)
img_LANCZOS = flat_image.resize((500, 400),resample=Image.LANCZOS)
img_HAMMING = flat_image.resize((500, 400),resample=Image.HAMMING)
img_BOX = flat_image.resize((500, 400),resample=Image.BOX)
plt.rcParams['xtick.bottom'] = plt.rcParams['xtick.labelbottom'] = False
plt.rcParams['xtick.top'] = plt.rcParams['xtick.labeltop'] = False
plt.rcParams['ytick.left'] = plt.rcParams['ytick.labelleft'] = False
plt.rcParams['ytick.right'] = plt.rcParams['ytick.labelright'] = False
# Show image using matplotlib
fig, ax = plt.subplots(2, 3, figsize=(16, 10))
# Remove space between subplots
fig.subplots_adjust(wspace=0.1, hspace=0)
# set title size
plt.rcParams['axes.titlesize'] = 20
ax[0, 0].imshow(img_NEAREST)
ax[0, 0].set_title('NEAREST')
ax[0, 1].imshow(img_BILINEAR)
ax[0, 1].set_title('BILINEAR')
ax[0, 2].imshow(img_BICUBIC)
ax[0, 2].set_title('BICUBIC')
ax[1, 0].imshow(img_LANCZOS)
ax[1, 0].set_title('LANCZOS')
ax[1, 1].imshow(img_HAMMING)
ax[1, 1].set_title('HAMMING')
ax[1, 2].imshow(img_BOX)
ax[1, 2].set_title('BOX')
plt.show()
# Save plot with transparent background
fig.savefig('resampling.png', transparent=True)


```
